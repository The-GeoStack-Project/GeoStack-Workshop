Nu willen een simpele visualisatie creeren. Hiervoor gaan we de Python libraries MatplotLib en Cartopy gebruiken.

```python
'''
Hier maken we een nieuwe plot door middel van MatplotLib
'''
plt.figure(figsize = (20, 12))

'''
Hier creeren we een nieuwe Cartopy map. 
'''
cartopyMap = plt.axes(projection=ccrs.PlateCarree())

'''
Hier voegen we de kustlijnen toe aan de Cartopy map
'''
cartopyMap.coastlines(resolution='10m')

'''
Hier creeren we het landopervlakte.
De het landoppervlakte wordt weergeven in het wit
De randen, van het landopervlakte worden weergeven in het zwart
'''
cartopyMap.add_feature(cartopy.feature.LAND.with_scale('10m'), edgecolor='black', facecolor = "white")

'''
Hier voegen we de zee toe aan de Cartopy map.
'''
cartopyMap.add_feature(cfeature.OCEAN) 

'''
Hier voegen we de meren toe aan de Cartopy map.
De randen van de meren worden weergeven in het zwart 
'''
cartopyMap.add_feature(cfeature.LAKES.with_scale('10m'), edgecolor = 'black')

'''
Hier voegen we het zee oppervlak toe aan de Cartopy map.
'''
cartopyMap.add_feature(cfeature.OCEAN) 

'''
Hier voegen we de rivieren toe aan de Cartopy map.
'''
cartopyMap.add_feature(cfeature.RIVERS.with_scale('10m')) 

'''
Hier voegen we de landsgrenzen toe aan de Cartopy map.
'''
cartopyMap.add_feature(cfeature.BORDERS.with_scale('10m'))


'''
Nu willen we de kraanvogel routes plotten op de Cartopy map, die we hierboven hebben gecreerd.
Dit doe we door middel van een scatter plot. Dit is een type plot van Matplotlib, 
voor het weergeven van datapunten.

Om een dataset te plotten moeten we eerst aangeven op welke map wel dat willen doen.
Daarna geven we aan wat voor plot we willen creeren.
Daarna moeten we een aantal waarde megeven aan de gekozen plot, deze waardes zijn als volgt:
Als eerste waarde geven we de longitude coordinaten mee.
Als tweede waarde geven we de latitude coordinaten mee.
Als derde waarde geven we aan wat voor kleur de kraanvogel route moet hebben.
Als vierde waarde geven we aan hoe groot elk datapunt moet worden.

In de line hieronder voegen we de kraanvogel: "Agnetha",toe aan de Cartopy map.
Eerst geven we aan dat we een plot willen weergeven op de cartopy map genaamd: "cartopyMap".
Daarna geven we het type plot aan genaamd: "scatter"
Daarna geven aan welke kolom de longitude coordinaten van de kraanvogel bevat. Dit doen we door eerst
aan te geven van welk dataframe we dit kolom willen selecteren, en dan de naam van de kolom mee te geven.
In dit geval is het dataframe: "Agnetha" en de naam van de kolom: "location-long".
Daarna doen we hetzelfde voor de latitude coordinaten.
Daarna geven we aan welke kleur de datapunten moeten krijgen, rood in dit geval. 
Daarna geven we aan hoe groot de datapunten moeten worden, 1 in dit geval.
'''
cartopyMap.scatter(Agnetha['location-long'], Agnetha['location-lat'],color="red", s = 1)

# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
#                                                                                       #
#             OPDRACHT 3: Voeg de kraanvogel:"Frida", toe aan de Cartopy Map.           #
#                                                                                       #
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # 
cartopyMap.scatter(Frida['location-long'], Frida['location-lat'],color="blue", s = 1)

'''
Hier voegen we de kraanvogel: "Casja",toe aan de Cartopy map.
'''
cartopyMap.scatter(Cajsa['location-long'], Cajsa['location-lat'],color="green", s = 1)
```
